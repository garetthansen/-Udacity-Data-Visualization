## Investigation Overview

> For this analysis, I chose to investigate several variables that I felt could potentially impact a borrower's Annual Percentage Rate (APR) for better or worse. These variables are income range, credit score (and rating), and loan size.

## Dataset Overview

> The provided data is from marketplace lending company Prosper and is comprised of 113,937 individual loans across a wide range of loan types (i.e. home improvement, auto, student, etc). While 81 different variables are attributed to each loan, only the subset of variables listed previously were analyzed.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe

loans = pd.read_csv('prosperLoanData.csv')
```

```python
#change APR to percentage format

loans['BorrowerAPR'] = loans['BorrowerAPR']*100
```

```python
#calculate average credit score

avg_credit_score = loans[['CreditScoreRangeLower', 'CreditScoreRangeUpper']].mean(axis=1)

#insert average credit score column after other credit score columns

loans.insert(loans.columns.get_loc('CreditScoreRangeUpper')+1, 'avg_credit_score', 
             avg_credit_score, allow_duplicates = False)
```

```python
# order min -> max

inc_ordered = ['Not displayed', 'Not employed', '$0', '$1-24,999', '$25,000-49,999', '$50,000-74,999', 
               '$75,000-99,999', '$100,000+']
```

```python
# create credit score bins

credit_bins = [300, 579, 669, 739, 799, 850]
credit_names = ['Very Poor', 'Fair', 'Good', 'Very Good', 'Exceptional']

# new credit_rating column

loans['credit_rating'] = pd.cut(loans['avg_credit_score'], credit_bins, labels = credit_names)
```

```python
# create bins for loan sizes

size_bins = [1000, 4000, 6500, 12000, 35000]
size_names = ['Very Small', 'Small', 'Medium', 'Large']

# create new loan size column

loans['loan_size'] = pd.cut(loans['LoanOriginalAmount'], size_bins, labels = size_names)
```

## Distribution of APR

Borrower rates span a wide range of possible values, hinting that the rate applied to a loan can widely vary borrower to borrower. The rates here take on a nearly unimodal distribution, with the notable exception around the 36% area.

```python
bins = np.arange(0, 45, 1)

plt.figure(figsize=[14.70, 8.27])
plt.hist(data = loans, x = 'BorrowerAPR', bins = bins, color = 'teal')
plt.title('Distribution of APR %', size = 20)
plt.ylabel('Count of Loans', size = 15)
plt.xlabel('APR %', size = 15);
```

## APR vs Income Range and Credit Rating

When factoring annual income, those with a favorable credit rating consistently experienced a lower APR than those with less favorable credit ratings; howeve, while those with good, very good, or exceptional ratings experienced a lower APR as income increased, those with fair or very poor credit ratings experienced slightly similar or occasionally slightly higher APR as income increased.

```python
plt.figure(figsize=[14.70, 8.27])
sb.pointplot(data = loans, x = 'IncomeRange', y = 'BorrowerAPR', hue = 'credit_rating', order = inc_ordered)
plt.title('APR % vs Income Range and Credit Rating', size = 20)
plt.xlabel('Income Range', size = 15)
plt.ylabel('APR %', size = 15)
plt.legend(bbox_to_anchor=(1, .89),
           bbox_transform=plt.gcf().transFigure);
```

## APR vs Income Range and Loan Size

Uemployed borrowers experience some of the highest APR rates, regardless of loan size. Very small loans (1000-4000) experience the highest rates for those with a provided income, although the rates do rise and lower slightly depending on income range. Small and medium-sized loans experience very similar APR rates for those with reported income and are nearly identical for those earning above 50,0000. Perhaps unsurprisingly, those with high incomes and large loan sizes enjoy some of the lowest rates.

```python
plt.figure(figsize=[14.70, 8.27])
sb.pointplot(data = loans, x = 'IncomeRange', y = 'BorrowerAPR', hue = 'loan_size', order = inc_ordered, 
             palette = 'viridis_r', dodge = .4, linestyles = "--")
plt.title('APR % vs Income Range and Loan Size', size = 20)
plt.xlabel('Income Range', size = 15)
plt.ylabel('APR %', size = 15)
plt.legend(bbox_to_anchor=(1, .89),
           bbox_transform=plt.gcf().transFigure);
```
